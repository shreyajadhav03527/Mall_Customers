```python
# elbow methods
# it helps to find how  many groups to use
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)


  # wcss-within cluster sum of squares
  # lower wcss-tightly clusters
  # higher wcss-less dense clusters

  # we will store this for each kmeans each loop new model different number of clusters

  # in each loop k increases ,wcss decreases and improvement slows down after certain point
```

```python
plt.plot(range(1,11),wcss)
plt.title('elbow method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
KMeans=KMeans(n_clusters=5,random_state=42)
label=KMeans.fit_predict(x)
```

```python
df["cluster"]=label
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'x' (annual_income_k and spending_score_1-100) is already defined and available
# We'll use the range of 'spending_score_1-100' for the x-axis for context
plt.figure(figsize=(10, 6))
plt.scatter(x['annual_income_k'], x['spending_score_1-100'], label='Customer Data')

# Plot y = 10 as a horizontal line
# Using the min/max of 'annual_income_k' to define the extent of the horizontal line
plt.axhline(y=10, color='r', linestyle='--', label='y = 10')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Data with y = 10 Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# applying PCA here
# principal compound analysis-pca
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x)

# components=2-reducing data 2 dimensions,information preserved mostliy
# PCA are used when data has many feature,so pca compress information then we visualize it
# pca does not create clusters,it helps us see them
```

```python
# visualize clusters using PCA
plt.scatter(x_pca[:,0],x_pca[:,1],c=df['cluster'])
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.title('customer segments')
plt.show()
```

```python
df.groupby('cluster')[['annual_income_k','spending_score_1-100']].mean()
```

```python
df['segment']=df['cluster'].map({
    0:'mid income- mid spend',
    1:'high income-high spend',
    2:'low income-high spend',
    3:'low income-low spend',
    4:'mid income-low spend'


})
```

```python
df.head()
```

```python
plt.figure(figsize=(5,5))
for segment_name,cluster_id in df[['segment','cluster']].drop_duplicates().values:
  subset=df[df['cluster']==cluster_id]

  plt.scatter(subset['annual_income_k'],subset['spending_score_1-100'],label=segment_name
  )

  plt.xlabel('annual_income_k')
  plt.ylabel('spending_score_1-100')
  plt.title('customer segments')
  plt.legend(title='customer segment')
  plt.show()
```

```python
plt.figure(figsize=(7,7))
for segment_name, group in df.groupby('segment'):
    plt.scatter(
        group['annual_income_k'],
        group['spending_score_1-100'],
        label=segment_name
    )

plt.xlabel('annual_income_k')
plt.ylabel('spending_score_1-100')
plt.title('customer segments')
plt.legend(title='customer segment')
plt.show()
```
